```python
#Dependencies

import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from datetime import datetime, timedelta
import csv
import matplotlib.dates as mdates
```

```python
#Data files
newyork2020_county_path = "Resources/newyorkcountydata2020.csv"

#Ready the data
newyork2020_county = pd.read_csv(newyork2020_county_path)

#Display
newyork2020_county.head()
```

```python
#Generate Line Plot of Total Homes Sold in 2020
plt.style.use('seaborn-whitegrid')
dates = newyork2020_county['period_begin']
dates = pd.to_datetime(dates)
plt.plot(dates,newyork2020_county['total_homes_sold'],color = 'red')
plt.gcf().autofmt_xdate()
plt.xlabel('Period')
plt.ylabel('Number of Homes Sold')
plt.title('Number of Homes Sold in New York County, NY in 2020')
plt.savefig("abigail/homessold2020.png")
plt.show()
```
